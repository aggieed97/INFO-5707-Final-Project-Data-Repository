# Conference

```python
def camel_to_snake(name):
    """
    Converts a camelCase string to snake_case.
    Used to replicate the functionality of R's janitor::clean_names().
    """
    # Insert an underscore before any capital letter followed by a lowercase one
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    # Insert an underscore before any remaining capital letters
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()

def fetch_and_parse_conferences():
    """
    Fetches men's college basketball conference data from the ESPN API,
    parses the JSON, and loads it into a pandas DataFrame, applying
    cleaning and renaming to match the provided R script's logic.
    """
    api_url = "http://site.api.espn.com/apis/site/v2/sports/basketball/mens-college-basketball/scoreboard/conferences?seasontype=2"
    
    print(f"Fetching data from: {api_url}\n")
    
    try:
        # 1. Fetch the data from the API
        response = requests.get(api_url)
        response.raise_for_status()  # Raise an exception for bad status codes (4xx or 5xx)
        
        # 2. Parse the JSON response
        data = response.json()
        
        # 3. Check if the expected 'conferences' key exists
        if 'conferences' in data and isinstance(data['conferences'], list):
            conferences_list = data['conferences']
            print(f"Successfully retrieved {len(conferences_list)} conference records.")

            # 4. Create a pandas DataFrame from the list of dictionaries
            df = pd.DataFrame(conferences_list)
            
            # --- R Code Equivalents Start Here ---

            # 5. janitor::clean_names() equivalent (camelCase to snake_case)
            df.columns = [camel_to_snake(col) for col in df.columns]
            
            # 6. dplyr::filter(!(.data$group_id %in% c(0,50))) equivalent
            # Note: groupId comes as a string, so we filter by string values '0' and '50'
            df_filtered = df[~df['group_id'].isin(['0', '50'])].copy()

            # 7. dplyr::mutate() equivalent (Type Conversion and New ID)
            # Create conference_id and convert IDs to integer types
            df_filtered['group_id'] = pd.to_numeric(df_filtered['group_id'], errors='coerce', downcast='integer')
            df_filtered['conference_id'] = df_filtered['group_id']
            df_filtered['parent_group_id'] = pd.to_numeric(
                df_filtered['parent_group_id'], errors='coerce', downcast='integer'
            )
            
            # 8. dplyr::select(-"subGroups") is implicitly handled by the final column selection.
            # Fill remaining NaN values (e.g., from converting missing strings to numeric)
            df_filtered = df_filtered.fillna('')

            # 9. dplyr::rename() equivalent
            rename_mapping = {
                "short_name": "conference_short_name",
                "uid": "conference_uid",
                "name": "conference_name",
                "logo": "conference_logo"
            }
            df_final = df_filtered.rename(columns=rename_mapping)
            
            # 10. Select final columns in desired order (Implicit in R's workflow)
            final_cols = [
                'conference_name', 
                'conference_short_name', 
                'conference_id',
                'group_id', 
                'conference_uid', 
                'parent_group_id', 
                'conference_logo'
            ]
            
            # Ensure only existing columns are selected to avoid KeyError (in case 'subGroups' was implicitly dropped)
            df_final = df_final[[col for col in final_cols if col in df_final.columns]]
            
            return df_final
        else:
            print("Error: 'conferences' key not found in the API response.")
            return pd.DataFrame() # Return empty DataFrame on failure

    except requests.exceptions.RequestException as e:
        print(f"An error occurred during the API request: {e}")
        return pd.DataFrame()
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return pd.DataFrame()

# Execute the function and display the result
conference_df = fetch_and_parse_conferences()

if not conference_df.empty:
    print("\n--- College Basketball Conferences DataFrame (R-Style) ---\n")
    # Print the full DataFrame without truncation for better viewing
    with pd.option_context('display.max_rows', None, 'display.max_columns', None):
        display(conference_df)
    
    # You can also save it to a CSV file if needed
    # conference_df.to_csv('mbb_conferences.csv', index=False)
```

```python
import requests
import pandas as pd
import re

# Set a default season year for the teams API (ESPN's APIs often require a season context)
CURRENT_SEASON_YEAR = 2026

def camel_to_snake(name):
    """
    Converts a camelCase string to snake_case.
    Used to replicate the functionality of R's janitor::clean_names().
    """
    # Insert an underscore before any capital letter followed by a lowercase one
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    # Insert an underscore before any remaining capital letters
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()

def get_conferences_df():
    """
    Fetches men's college basketball conference data, parses the JSON, and 
    returns it as a pandas DataFrame after cleaning (matching R script logic).
    (Renamed from fetch_and_parse_conferences for modularity)
    """
    api_url = "http://site.api.espn.com/apis/site/v2/sports/basketball/mens-college-basketball/scoreboard/conferences?seasontype=2"
    
    try:
        response = requests.get(api_url)
        response.raise_for_status()
        data = response.json()
        
        if 'conferences' in data and isinstance(data['conferences'], list):
            df = pd.DataFrame(data['conferences'])
            
            # 1. janitor::clean_names() equivalent
            df.columns = [camel_to_snake(col) for col in df.columns]
            
            # 2. dplyr::filter(!(.data$group_id %in% c(0,50))) equivalent
            df_filtered = df[~df['group_id'].isin(['0', '50'])].copy()

            # 3. dplyr::mutate() equivalent (Type Conversion and New ID)
            df_filtered['group_id'] = pd.to_numeric(df_filtered['group_id'], errors='coerce', downcast='integer')
            df_filtered['conference_id'] = df_filtered['group_id']
            df_filtered['parent_group_id'] = pd.to_numeric(
                df_filtered['parent_group_id'], errors='coerce', downcast='integer'
            )
            
            # 4. Fill remaining NaN values (e.g., from converting missing strings to numeric)
            df_filtered = df_filtered.fillna('')

            # 5. dplyr::rename() equivalent
            rename_mapping = {
                "short_name": "conference_short_name",
                "uid": "conference_uid",
                "name": "conference_name",
                "logo": "conference_logo"
            }
            df_final = df_filtered.rename(columns=rename_mapping)
            
            # 6. Select final columns in desired order
            final_cols = [
                'conference_name', 'conference_short_name', 'conference_id',
                'group_id', 'conference_uid', 'parent_group_id', 'conference_logo'
            ]
            
            df_final = df_final[[col for col in final_cols if col in df_final.columns]]
            
            return df_final
        else:
            print("Error: 'conferences' key not found in the API response.")
            return pd.DataFrame()

    except requests.exceptions.RequestException as e:
        print(f"An error occurred during conference API request: {e}")
        return pd.DataFrame()
    except Exception as e:
        print(f"An unexpected error occurred during conference parsing: {e}")
        return pd.DataFrame()

# --- Helper Functions for Teams Parsing ---

def flatten_logos(logos_list):
    """
    Helper function to mimic tidyr::unnest_wider for the logos array.
    Extracts the first two logo URLs for a team.
    """
    if not isinstance(logos_list, list) or not logos_list:
        return {'logo': None, 'logo_dark': None}
    
    # logo is logos_href_1 (first item's href)
    logo = logos_list[0].get('href') if len(logos_list) > 0 else None
    
    # logo_dark is logos_href_2 (second item's href)
    logo_dark = logos_list[1].get('href') if len(logos_list) > 1 else None
    
    return {'logo': logo, 'logo_dark': logo_dark}

def extract_summary_record(record_data):
    """
    Helper to extract the main record summary.
    Replicates the primary goal of the complex R unnesting/pivoting 
    which often focuses on the high-level summary.
    """
    if isinstance(record_data, dict) and 'items' in record_data and isinstance(record_data['items'], list):
        # We look for the first item, which usually holds the overall record
        first_item = record_data['items'][0]
        return first_item.get('summary', '')
    return ''

def get_conference_teams_map(year):
    """
    Performs the multi-API calls to map teams to their specific conference group ID.
    """
    print("Starting multi-API calls to map teams to conferences...")
    base_url = "http://sports.core.api.espn.com/v2/sports/basketball/leagues/mens-college-basketball/seasons"
    # This URL fetches links to all the conferences (groups/50/children)
    conferences_base_url = f"{base_url}/{year}/types/2/groups/50/children?limit=1000&lang=en&region=us"

    try:
        # First call: Get links to all 32 conference groups
        res_conf_links = requests.get(conferences_base_url)
        res_conf_links.raise_for_status()
        
        conf_data = res_conf_links.json()
        conf_items = conf_data.get("items", [])
        
        # Extract the href and calculate the group_id for each conference
        conf_links_df = pd.DataFrame(conf_items)
        if conf_links_df.empty:
            print("Warning: No conference group links found.")
            return pd.DataFrame()
            
        conf_links_df = conf_links_df.rename(columns={"$ref": "href"})
        # Extract group_id from the href string using regex
        conf_links_df['group_id'] = conf_links_df['href'].str.extract(r"groups\/(\d+)")
        
        # Construct the team list URL for each conference
        conf_links_df['teams_url'] = (
            base_url + "/" + 
            str(year) + 
            "/types/2/groups/" + 
            conf_links_df['group_id'] + 
            "/teams?lang=en&region=us"
        )
        
        all_conference_teams = []
        
        # Second series of calls: Iterate through each conference to get its teams
        for index, row in conf_links_df.iterrows():
            teams_url = row['teams_url']
            group_id = row['group_id']
            
            try:
                res_teams = requests.get(teams_url)
                res_teams.raise_for_status()
                teams_data = res_teams.json()
                
                # The teams are under 'items' in this response
                teams_items = teams_data.get('items', [])
                if teams_items:
                    temp_df = pd.DataFrame(teams_items)
                    temp_df = temp_df.rename(columns={"$ref": "href"})
                    
                    # Extract team_id from the href string
                    temp_df['team_id'] = temp_df['href'].str.extract(r"teams\/(\d+)")
                    temp_df['group_id'] = pd.to_numeric(group_id, errors='coerce', downcast='integer')
                    
                    all_conference_teams.append(temp_df)
            except requests.exceptions.RequestException as e:
                print(f"Skipping group ID {group_id} due to API error: {e}")
                continue

        if all_conference_teams:
            # Combine all team lists into one DataFrame
            conference_teams_df = pd.concat(all_conference_teams, ignore_index=True)
            
            # FIX: Convert team_id to integer type to match the 'teams' DataFrame for merging
            conference_teams_df['team_id'] = pd.to_numeric(
                conference_teams_df['team_id'], 
                errors='coerce', 
                downcast='integer'
            )
            
            return conference_teams_df[['team_id', 'group_id']]
        else:
            print("No team data successfully retrieved from conference groups.")
            return pd.DataFrame()

    except requests.exceptions.RequestException as e:
        print(f"An error occurred during conference mapping setup: {e}")
        return pd.DataFrame()


def fetch_and_parse_teams(year=CURRENT_SEASON_YEAR):
    """
    Fetches, parses, cleans, and joins all college basketball team data
    with conference information, mimicking the R script workflow.
    """
    teams_base_url = "http://site.api.espn.com/apis/site/v2/sports/basketball/mens-college-basketball/teams?limit=1000"
    
    print(f"Fetching team data from: {teams_base_url}")
    
    try:
        # 1. Fetch the primary team data
        response = requests.get(teams_base_url)
        response.raise_for_status()
        data = response.json()

        # 2. Extract the nested list of teams
        # teams are deeply nested: sports -> leagues -> teams[0] -> team
        teams_list = data['sports'][0]['leagues'][0]['teams']
        
        # The actual team list is under the 'team' key of the first item
        leagues_data = [item['team'] for item in teams_list]
        leagues = pd.DataFrame(leagues_data)
        
        # 3. Logo Unnesting equivalent: Apply the helper function to flatten 'logos'
        logo_data = leagues['logos'].apply(flatten_logos).apply(pd.Series)
        leagues = pd.concat([leagues, logo_data], axis=1)
        
        # 4. Record/Stats Unnesting equivalent: Extract the summary field
        if 'record' in leagues.columns:
            leagues['summary'] = leagues['record'].apply(extract_summary_record)
            leagues = leagues.drop(columns=['record'])
            
        # 5. Drop unnecessary columns
        league_drop_cols = ["links", "isActive", "isAllStar", "uid", "slug"]
        leagues = leagues.drop(columns=[col for col in league_drop_cols if col in leagues.columns])
        
        # 6. Renaming columns
        rename_mapping = {
            "id": "team_id",
            "location": "team",
            "name": "mascot",
            "shortDisplayName": "short_name",
            "displayName": "display_name",
            "alternateColor": "alternate_color",
        }
        teams = leagues.rename(columns=rename_mapping)
        
        # 7. Mutate/Type conversion
        teams['team_id'] = pd.to_numeric(teams['team_id'], errors='coerce', downcast='integer')
        
        # 8. Fetch and join conference data (from the R script's function call)
        conferences_df = get_conferences_df()
        
        # 9. Fetch conference mapping (Multi-call process)
        conference_teams = get_conference_teams_map(year)

        # 10. Left Joins
        # Join team data with the group_id mapping
        teams_final = teams.merge(
            conference_teams, 
            on='team_id', 
            how='left'
        )
        
        # Join team data (now with group_id) with the conference details
        teams_final = teams_final.merge(
            conferences_df[[
                'group_id', 
                'conference_name', 
                'conference_short_name', 
                'conference_logo'
            ]], 
            on='group_id', 
            how='left'
        )

        # Final cleaning and ordering
        teams_final = teams_final.fillna('')
        
        # Optional: Apply janitor::clean_names() style to remaining columns (like color, alternate_color)
        teams_final.columns = [camel_to_snake(col) for col in teams_final.columns]
        
        # Select and reorder final output columns for clarity
        final_cols = [
            'team_id', 'team', 'mascot', 'display_name', 'short_name',
            'conference_name', 'group_id', 'summary', 
            'color', 'alternate_color', 'logo', 'logo_dark'
        ]
        
        return teams_final[[col for col in final_cols if col in teams_final.columns]]

    except requests.exceptions.RequestException as e:
        print(f"An error occurred during primary team API request: {e}")
        return pd.DataFrame()
    except Exception as e:
        print(f"An unexpected error occurred during team parsing: {e}")
        return pd.DataFrame()

# Execute the teams function and display the result
conference_df = get_conferences_df() # Run the original function to show it still works
teams_df = fetch_and_parse_teams()

if not teams_df.empty:
    print("\n--- College Basketball Teams DataFrame (R-Style Equivalent) ---\n")
    # Print the full DataFrame without truncation for better viewing
    # with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.width', 1000):
    #     print(teams_df)
    
    # You can also save it to a CSV file if needed
    # teams_df.to_csv('mbb_teams.csv', index=False)
```
